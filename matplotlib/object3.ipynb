```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4],[471, 481, 532, 537])
font1 = {'color':'b', 'weight':'bold', 'size':14}
font2 = {'color':'r', 'size':14}

plt.xlabel("Turn", labelpad=15, fontdict=font1, loc='right')
plt.ylabel("tasu", labelpad=20, fontdict=font2, loc='top')


plt.show()
```

```python
plt.plot([1, 2, 3, 4],[471, 481, 532, 537], '*-', color = 'k', label = 'Kim**')
plt.plot([1, 2, 3, 4],[410, 392, 398, 422], linestyle='-.', color = 'c', label = 'Lee**')
plt.plot([1, 2, 3, 4],[240, 270, 291, 354], linestyle=(0,(5,1)), color = 'violet', label = 'Park**')
plt.plot([1, 2, 3, 4],[701, 744, 728, 756], linestyle=(0,(3,5,1,5)), color = 'lavender', label = 'You**')

font1 = {'color':'b', 'weight':'bold', 'size':14}
font2 = {'color':'r', 'size':14}
plt.legend(loc=(1,1), ncol=2, fontsize=10, shadow=False)

plt.xlabel("Turn", labelpad=15, fontdict=font1, loc='right')
plt.ylabel("tasu", labelpad=20, fontdict=font2, loc='top')


plt.show()
```

```python
plt.plot([1, 2, 3, 4],[471, 481, 532, 537], linewidth="10", solid_capstyle="butt", label = 'Kim**')
plt.plot([1, 2, 3, 4],[410, 392, 398, 422], linewidth="7", solid_capstyle="round", label = 'Lee**')
plt.plot([1, 2, 3, 4],[240, 270, 291, 354], "bo-", label = 'Park**')
plt.plot([1, 2, 3, 4],[701, 744, 728, 756], "cx-", label = 'You**')

plt.legend(loc=(1,1), ncol=2, fontsize=10, shadow=False)

plt.xlabel("Turn", labelpad=15, fontdict=font1, loc='right')
plt.ylabel("tasu", labelpad=20, fontdict=font2, loc='top')


plt.show()
```
